```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Sample data (Area in acres, Rainfall in mm)
X = np.array([
    [1000, 200],
    [1500, 250],
    [1200, 300],
    [3000, 100],
    [3200, 120],
    [3500, 150]
])
```

```python
kmeans_biased = KMeans(n_clusters=2, random_state=0)
labels_biased = kmeans_biased.fit_predict(X)

print("Biased cluster labels:", labels_biased)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans_unbiased = KMeans(n_clusters=2, random_state=0)
labels_unbiased = kmeans_unbiased.fit_predict(X_scaled)

print("Unbiased cluster labels:", labels_unbiased)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(X[:,0], X[:,1], c=labels_biased)
plt.title("Biased K-Means")
plt.xlabel("Area")
plt.ylabel("Rainfall")

plt.subplot(1,2,2)
plt.scatter(X[:,0], X[:,1], c=labels_unbiased)
plt.title("Unbiased K-Means")
plt.xlabel("Area")
plt.ylabel("Rainfall")

plt.show()
```
